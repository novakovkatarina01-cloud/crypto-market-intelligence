```python
import requests
import pandas as pd

url = (
    "https://api.coingecko.com/api/v3/coins/markets"
    "?vs_currency=usd"
    "&order=market_cap_desc"
    "&per_page=10"
    "&page=1"
)

response = requests.get(url)

data = response.json()

df = pd.DataFrame(data)

df.head()
```

```python
crypto_df = df[
    [
        "id",
        "symbol",
        "name",
        "current_price",
        "market_cap",
        "market_cap_rank",
        "total_volume",
        "high_24h",
        "low_24h",
        "price_change_percentage_24h",
        "last_updated"
    ]
]

crypto_df.head()
```

```python
crypto_df.isnull().sum()
```

```python
from datetime import datetime

crypto_df["etl_load_timestamp"] = datetime.now()

crypto_df.head()
```

```python
spark_df = spark.createDataFrame(crypto_df)

spark_df.show()
```

```python
import matplotlib.pyplot as plt

top_market_cap = crypto_df.sort_values(
    by="market_cap",
    ascending=False
)

plt.figure(figsize=(10, 5))
plt.bar(
    top_market_cap["symbol"],
    top_market_cap["market_cap"]
)

plt.title("Top Cryptocurrencies by Market Cap")
plt.xlabel("Symbol")
plt.ylabel("Market Cap")

plt.show()
```

```python
import matplotlib.pyplot as plt

top_price = crypto_df.sort_values(
    by="current_price",
    ascending=False
)

plt.figure(figsize=(10, 5))

plt.bar(
    top_price["symbol"],
    top_price["current_price"]
)

plt.title("Top Cryptocurrencies by Current Price")
plt.xlabel("Symbol")
plt.ylabel("Current Price (USD)")

plt.show()
```

```python
top_volume = crypto_df.sort_values(
    "total_volume",
    ascending=False
)

plt.figure(figsize=(10,6))
plt.bar(
    top_volume["symbol"],
    top_volume["total_volume"]
)

plt.title("Trading Volume by Cryptocurrency")
plt.xlabel("Symbol")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    crypto_df["symbol"],
    crypto_df["price_change_percentage_24h"]
)

plt.axhline(y=0, color="red")

plt.title("24h Price Change (%)")
plt.xlabel("Symbol")
plt.ylabel("% Change")
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.scatter(
    crypto_df["market_cap"],
    crypto_df["current_price"],
    s=crypto_df["total_volume"] / 10000000,
    alpha=0.6
)

for i, row in crypto_df.iterrows():
    plt.text(
        row["market_cap"],
        row["current_price"],
        row["symbol"]
    )

plt.xlabel("Market Cap")
plt.ylabel("Current Price")
plt.title("Crypto Ecosystem Bubble Chart")

plt.show()
```

```python
import seaborn as sns

numeric_cols = [
    "current_price",
    "market_cap",
    "total_volume",
    "high_24h",
    "low_24h",
    "price_change_percentage_24h"
]

plt.figure(figsize=(10,8))

sns.heatmap(
    crypto_df[numeric_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Crypto Market Correlation Heatmap")
plt.show()
```
